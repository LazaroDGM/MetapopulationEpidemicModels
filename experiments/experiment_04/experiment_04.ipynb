```python
from models.estimation_model import DE_ESTIMATOR_SIR_EULER, DE_ESTIMATOR_SIR_LAGRANGE
import numpy as np
from sklearn.metrics import mean_squared_error
from time import time
```

```python
def GenerateParams_SIR_EULER(K):

    F = np.random.rand(K,K)
    np.fill_diagonal(F,0)
    F = F / K
    Beta = np.random.rand(K)
    Gamma = np.random.rand(K)
    return F, Beta, Gamma

def GenerateY0_SIR_EULER(K):
    y0 = np.zeros((4,K))
    y0[3] = np.round(np.random.rand(K) * 5000 + 1000, 0)
    y0[1] = np.round(np.random.rand(K) * 5 + 1)
    y0[0] = y0[3] - y0[1]
    y0.resize((4*K,))
    return y0

def GenerateParams_SIR_LAGRANGE(K):

    Out = np.random.rand(K,K)
    np.fill_diagonal(Out,0)
    Out = Out / K

    In = np.random.rand(K,K)
    np.fill_diagonal(In,0)
    In = In / K

    Beta = np.random.rand(K)
    Gamma = np.random.rand(K)

    return Out, In, Beta, Gamma

def GenerateY0_SIR_LAGRANGE(K):
    y0 = np.zeros((4,K,K))
    np.fill_diagonal(y0[3], np.round(np.random.rand(K) * 5000 + 1000, 0))
    np.fill_diagonal(y0[1], np.round(np.random.rand(K) * 5 + 1))
    y0[0] = y0[3] - y0[1]
    y0.resize((4*K*K,))
    return y0
```

```python
test_1_params = GenerateParams_SIR_EULER(16)
test_1_y0 = GenerateY0_SIR_EULER(16)
```

```python
test_1_estimator = DE_ESTIMATOR_SIR_EULER(16)
test_1_estimator.set_initial_value(0, test_1_y0)
test_1_estimator.set_params(test_1_params)
```

```python
test_1_result = test_1_estimator.estimate_with_model(301)
```
